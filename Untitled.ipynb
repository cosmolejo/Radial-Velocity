```python
###################################
## LOAD DATA
##################################
import numpy as np

data = np.loadtxt('RV_dataset_2.dat.txt')

HJD= data[:,0]
RV = data[:,1]
RV_err=data[:,2]
```

```python
####################################
##PLOT DATA
####################################

import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(HJD,RV,'*--',label='RV')
plt.legend()
plt.show()
```

```python
###############################
## lombscargle
###############################

import scipy.signal as signal
nout = 100000
normval = HJD.shape[0] # For normalization of the periodogram
f = np.linspace(0.01, 10, nout)
pgram = signal.lombscargle(HJD, RV, f)

period=2*np.pi/f[pgram==max(pgram)]
print (period)
#Now make a plot of the input data:
plt.figure(figsize=(20,10))
plt.subplot(2, 1, 1)
plt.plot(HJD % period, RV, '.') #b+--')

#Then plot the normalized periodogram:

plt.subplot(2, 1, 2)

plt.plot((2*np.pi)/f, np.sqrt(4*(pgram/normval)))
plt.xticks(np.arange(2, 2+1, 0.01))
plt.xlim(2.1,2.3)
plt.grid()
plt.show()
```

```python
##################################
## FIT
##################################
w=2*np.pi/period
#x=np.linspace(HJD[0],HJD[-1],100)
x=np.linspace(0,period,100)


def fit (x):
    return 240*np.sin((w*x)+3.4)


 
y= fit(x)
err=RV-fit(HJD)


plt.figure(figsize=(20,10))
plt.plot(HJD %period, RV, 'o',label="RV obs")
plt.plot(HJD %period,err,'go',label="RV err")
#plt.errorbar(HJD , RV, yerr=RV_err, fmt='.')
plt.plot(x,y,'r--', label="RV fit")
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import leastsq
import pylab as plt

N = 1000 # number of data points
t = np.linspace(0, 4*np.pi, N)
data = 3.0*np.sin(t+0.001) + 0.5 + np.random.randn(N) # create artificial data with noise

guess_mean = np.mean(data)
guess_std = 3*np.std(data)/(2**0.5)
guess_phase = 0

# we'll use this to plot our first estimate. This might already be good enough for you
data_first_guess = guess_std*np.sin(t+guess_phase) + guess_mean

# Define the function to optimize, in this case, we want to minimize the difference
# between the actual data and our "guessed" parameters
optimize_func = lambda x: x[0]*np.sin(t+x[1]) + x[2] - data
est_std, est_phase, est_mean = leastsq(optimize_func, [guess_std, guess_phase, guess_mean])[0]

# recreate the fitted curve using the optimized parameters
data_fit = est_std*np.sin(t+est_phase) + est_mean

plt.plot(data, '.')
plt.plot(data_fit, label='after fitting')
plt.plot(data_first_guess, label='first guess')
plt.legend()
plt.show()
```
